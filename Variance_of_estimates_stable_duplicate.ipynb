# **Parameter Estimation of the Tail Index of Symmetric Stable Distributions**

**Abstract**

In this report, we demonstrate how to estimate the tail index of stable distribution $S_{\alpha}(1,0,0)$ using the MLP method. The theory behind this script is based on the paper [Parameter Estimation of Sub-Gaussian Stable Distributions](https://eudml.org/doc/262151) by Omelchenko V. (2014). In this script, the author extends this method and provides a more compact formulation of it. The paper focuses on the estimation of the parameter Î± of stable distributions and the parameters $\alpha$ and $\mathbf{R}$ of subgaussian distributions. In this report and its follow-ups, it is emphasized that this method can be applied to a much broader family of distributions.

**Key words**

Maximum Likelihood Projection (MLP), MLP estimates, orthogonal projections, projection vectors, characteristic function, empirical characteristic function, error function, Fisher information.

**Notation:**



*   A sample of observations $(X_1, X_2, ... ,X_n)$ is denoted $\mathbf{X}$.
*   The ***characteristic function*** of a random variable $X$ which depends on the parameter $\alpha$ ($X=X(\alpha)$) is denoted and defined $$\psi(t, \alpha)=\mathbb{E}\left(\exp(i\cdot t\cdot X )\right).$$ Or in other words: $\psi(t, \alpha)=\mathbb{E}\left(\exp(i\cdot t\cdot X(\alpha) )\right)$.
*   For a sample $\mathbf{X}$, the empirical characteristic function is denoted and defined $$\psi_{\mathbf{X}}(t)=\frac{1}{n}\sum_{j=1}^n\exp(i\cdot t\cdot X_j).$$ 
*   If $\mathbf{X}$ is from a distribution which is centered and symmetric, then $\psi(t)=\mathrm{Re}(\psi(t))$. In this case
$$\psi(t)=\mathbb{E}(\cos(t\cdot X)) \text{ and }\psi_{\mathbf{X}}(t)=\frac{1}{n}\sum_{j=1}^n\cos(t\cdot X_j).$$
*   A sample of points $t_1, t_2,...,t_k$ from the domain of the characteristic function $\psi(t)$ is denoted $\mathbf{T}$.
*   The method is called **Maximum Likelihood Projections (MLP)** therefore, the estimate is denoted $\alpha_{MLP}$. Its latest update is called **Stable Maximum Likelihood**.


**The formula of estimating the parameter $\alpha$ of the stable distribution**

\begin{align*}
  \hat{\alpha}_{_{MLP}}& =\mathrm{argmin}_{\alpha}\left| \left(  (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top} \cdot \psi_{\mathbf{X}}(\mathbf{T}) \right| \\
& & \\
%J^*_k(x,\alpha) &= 2a^*_0 + 2\sum_{j=1}^{k}a_j\cos(t_j x) \\
\mathbf{T} & = \{t_0, t_1, t_2, ..., t_k \}\text{ with }t_0=0,\\
\mathbf{T}{\oplus} \mathbf{T} & =  \{t_i+t_j,\quad i,j=0,1,..,k\}, \text{it is not the Kronecker sum}, \\
\mathbf{T}{\ominus} \mathbf{T} & =  \{t_i-t_j,\quad i,j=0,1,..,k\}, \\
\psi_{\mathbf{X}}(\mathbf{T}) &= \{1, \psi_{\mathbf{X}}(t_1), \psi_{\mathbf{X}}(t_2), ..., \psi_{\mathbf{X}}(t_k) \}, \quad \quad (\psi_{\mathbf{X}}(0)=1),\\
\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)) & = \{\psi(t_i+t_j),\quad i,j=0,1,...,k\}\text{ is a matrix to be inverted.}
\end{align*}

Since $\alpha_{_{MLP}}$ depends on $\mathbf{T}$, we can denote an estimate yielded from $\mathbf{T}$ as $\alpha_{_{MLP}}(\mathbf{T})$. Denoting $\mathbf{T}_\psi$ as the domain of $\psi(t)$, and the maximum likelihood estimate as $\alpha_{_{ML}}$, we have: 
$$\lim_{\mathbf{T}\rightarrow \mathbf{T}_\psi}\alpha_{_{MLP}}(\mathbf{T})=\alpha_{_{ML}}.$$

# **Explanation of the formula**


The Fisher information $I(\alpha)$ of the tail index $\alpha$ is calculated as follows:

if $X\sim S_{\alpha}(1,0,0)$ and $p(x,\alpha)$ is its density function, then
$$I(\alpha) = \int_{-\infty}^{+\infty}J^2(x, \alpha)p(x,\alpha)dx,$$
where $p(x,\alpha)$ is the density function of $X$ and $J(x,\alpha)=\frac{  \left( \frac{\partial p(x,\alpha)}{\partial \alpha} \right) }{p(x,\alpha)}$. 

For the sample $X_1, X_2, ..., X_n$ ***maximum likelihood*** estimate is obtained as follows:
$$\hat{\alpha}_{ML}=\left\{\alpha : \sum_{j=1}^{n} J(X_j, \alpha) = 0\right\}$$

If a proper estimate of the function $J(x, \alpha)$ is found, then it can be used for the estimation of the parameters. 

In the ***maximum likelihood projections*** method (MLP), the function $J(x,\alpha)$ is projected to the space $\{1, \exp(i\cdot t_1 x), \exp(i\cdot t_2 x), ..., \exp(i \cdot t_k x) \}$, i.e.
$$J_k(x,\alpha) = \sum_{j=0}^{k}a_j \exp(i\cdot t_jx),$$
where $t_0=0$, and $t_1,t_2,...,t_k$ are randomly chosen non-zero numbers that belong to the domain of the characteristic function. Their choice affects the quality of the estimates. We denote this finite and discrete set of points as follows:

$$\mathbf{T}=\{t_i, i=0,1,...,k \}.$$

The scalar product in this space is defined as follows:
if $X\sim S_{\alpha}(1,0,0)$, i.e. the characteristic function of $X$ is $\psi(t)=\exp\left( -|t|^{\alpha} \right)$, 

$$\langle \exp(i\cdot t_m X), \exp(i\cdot t_n X) \rangle =\mathbb{E}\exp(i\cdot t_m X)\cdot \exp(i\cdot t_n X)=$$
$$=\mathbb{E}\exp(i\cdot (t_m + t_n)\cdot X) = \exp(-|t_m+t_n|^{\alpha})=\psi(t_m+t_n). $$

The usage of orthogonal projections implies:
$$\langle (J_k(X,\alpha)-J(X,\alpha)), \exp(i\cdot t_j X)\rangle = 0,\quad j=1,2,...,k$$
which is equivalent to:
$$\sum_{i=0}^{k}a_i \exp(-|t_i+t_j|^{\alpha}) = \frac{\partial}{\partial \alpha}\exp(-|t_j|^{\alpha}),\quad j=1,2,...,k.$$


This will be proven for a general characteristic function:

\begin{align*}
\langle (J_k(X,\alpha)-J(X,\alpha)), \exp(i\cdot t_j X)\rangle &= 0 \\
\langle J_k(X,\alpha), \exp(i\cdot  t_j X)\rangle  &= \langle J(X,\alpha), \exp(i\cdot t_j X)\rangle  \\
\text{Transform both parts of the last equality} &  \\
\text{The left part:} &  \\
\langle J_k(X,\alpha), \exp(i\cdot  t_j X)\rangle & = \left\langle \sum_{i=0}^{k}a_i \exp(t_iX), \exp(i\cdot  t_j X)\right\rangle = \\
= \sum_{i=0}^k a_i \langle \exp(i\cdot t_i X), \exp(i\cdot t_j X)\rangle & = \sum_{i=0}^k a_i \cdot \psi(t_i+t_j).\\
\text{The right part:} &  \\
\langle J(X,\alpha), \exp(i\cdot t_j X)\rangle & = \int_{-\infty}^{\infty}J(x,\alpha)p(x,\alpha)\exp(i\cdot t_jx)dx=\\
\int_{-\infty}^{\infty} \frac{\left(  \frac{\partial p(x,\alpha)}{\partial \alpha} \right)}{p(x,\alpha)}p(x,\alpha)\exp(i\cdot t_jx)dx & = \int_{-\infty}^{\infty} \frac{\partial p(x,\alpha)}{\partial \alpha} \exp(i\cdot t_jx)dx = \\
\frac{\partial }{\partial \alpha}\int_{-\infty}^{\infty} p(x,\alpha) \exp(i\cdot t_jx)dx  & = \frac{\partial }{\partial \alpha}\psi(t_j) = \psi_{\alpha}(t_j).  
\end{align*}
The order of the integration and partial derivative can be reversed because of the Leibnitz rule and the fact that $p(x,\alpha)\geq 0$.


Thus, for a general characteristic function, this equation can be written as:
$$\sum_{i=0}^{k}a_i \psi(t_i+t_j,\alpha) = \frac{\partial}{\partial \alpha}\psi(t_j,\alpha), \quad j=1,2,...,k,$$
and for this specific case, we have the following system of equations:
$$\sum_{i=0}^{k}a_i \exp(-|t_i+t_j|^{\alpha}) = -|t_j|^{\alpha}\ln(|t_j|)\exp(-|t_j|^{\alpha}), \quad j=1,2,...,k.$$

If $A(\alpha) = \{\exp(-|t_i+t_j|^{\alpha}),\quad i,j=0,1,..,k\}$ and $b(\alpha)=\{0\}\cup \{  -|t_j|^{\alpha}\ln(|t_j|)\exp(-|t_j|^{\alpha}),\quad j=1,2,..,k \}$, then
$$J_k(x,\alpha) = \left( (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top} F(\mathbf{T}, X_j),$$
where $F(\mathbf{T}, X)=\{1, \exp(i\cdot t_1 X), \exp(i\cdot t_2 X), ..., \exp(i \cdot t_k X) \}$, and $\mathbf{T}=\{t_0,t_1,...,t_k\}$.

To get rid of the imaginary component, we can use the following trick associated with the choice of the parameters $t_1,t_2,...,t_k$:
$$J^*(x,\alpha)=\sum_{j=-k}^{k}a_j\exp\left( i\cdot t_jx\right) \text{ with }a_{-j}=a_j\text{ and }t_{-j}=t_j,j=1,2,...,k, $$
then 
$$J^*(x,\alpha) = a_0 + \sum_{j=1}^{k}a_j\left(\exp(i\cdot t_j x) +\exp(-i\cdot t_jx) \right) = a_0 + 2\sum_{j=1}^{k}a_j\cos(t_j x).$$

We can use the trick $a_0^* = a_0/2$, and have
$$J^*(x,\alpha) = 2a_0^* + 2\sum_{j=1}^{k}a^*_j\cos(t_j x)=2\sum_{j=0}^{k}a^*_j\cos(t_j x),\quad a_j^* = a_j\text{ if }j>0\text{ and }\frac{a_0}{2}\text{ if }j=0.$$

The corresponding system of equations in this formulation is 
$$\sum_{i=0}^{k}a_i \left( \exp(-|t_i+t_j|^{\alpha})  + \exp(-|t_i-t_j|^{\alpha})  \right)= -|t_j|^{\alpha}\ln(|t_j|)\exp(-|t_j|^{\alpha}), \quad j=1,2,...,k,$$
and $F(\mathbf{T}, X)=\{1, 2\cos(t_1 X), 2\cos(t_2 X), ..., 2\cos(t_k X) \}$ because

$$\langle J^*(X,\alpha), \exp(i\cdot t_i\cdot X) \rangle = \left\langle \sum_{i=0}^{k} a^*_j\left(\exp(i\cdot t_i X) +\exp(-i\cdot t_i X) \right), \exp(i\cdot t_j\cdot X)  \right\rangle=$$
$$= \sum_{i=0}^{k} a^*_j(\langle \exp(i\cdot t_i X),\exp(i\cdot t_j\cdot X) \rangle + \langle\exp(-i\cdot t_i X),\exp(i\cdot t_j\cdot X) \rangle) =  
$$
$$=\sum_{i=0}^{k}a_i^* \left( \exp(-|t_i+t_j|^{\alpha})  + \exp(-|t_i-t_j|^{\alpha})\right).$$

**Implementation in code:**

By replacing $a_0$ with $a^*_0 = a_0/2$, we get:

\begin{align*}
J^*_k(x,\alpha) &= 2a^*_0 + 2\sum_{j=1}^{k}a_j\cos(t_j x) \\
F(\mathbf{T}, X) &= \{2, 2\cos(t_1 X), 2\cos(t_2 X), ..., 2\cos(t_k X) \} \\
A(\alpha) & =  \{\exp(-|t_i+t_j|^{\alpha})+\exp(-|t_i-t_j|^{\alpha}),\quad i,j=0,1,..,k\} \\
b(\alpha) & = \{0\}\cup \{  -|t_j|^{\alpha}\ln(|t_j|)\exp(-|t_j|^{\alpha}),\quad j=1,2,..,k \}\\
\mathbf{Hence} & \\
J^*_k(x,\alpha) &= 2a^*_0 + 2\sum_{j=1}^{k}a_j\cos(t_j x) =  \left( (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top} F(\mathbf{T}, x) \\
\sum_{j=1}^{n}J^*_k(X_m,\alpha)& = \left( (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top} \sum_{m=1}^{n}F(\mathbf{T}, X_m) = 2\left(  (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top}  \psi_{\mathbf{X}}(\mathbf{T})\\
\mathbf{Finally} & \\
 \hat{\alpha}_{_{MLP}}& =\mathrm{agrmin}_{\alpha}\left| \left( (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top} \frac{1}{n}\sum_{m=1}^{n}F(\mathbf{T}, X_m) \right| \\
\text{or} & \\
  \hat{\alpha}_{_{MLP}}& =\mathrm{agrmin}_{\alpha}\left| \left(  (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top}  \psi_{\mathbf{X}}(\mathbf{T}) \right| \\
\text{or} & \\
\hat{\alpha}_{_{MLP}}& =\operatorname{agrmin}_{\alpha}\left| \left(  (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)+\psi(\mathbf{T}{\ominus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top} \cdot \psi_{\mathbf{X}}(\mathbf{T}) \right| 
\end{align*}

Description of the functions from the following cell:

-  **A(a, T)** is used to generate the matrix $A(\alpha) = \{\exp(-|t_i+t_j|^{\alpha}) + \exp(-|t_i-t_j|^{\alpha}),\quad i,j=0,1,..,k\}$
-  **b(a, T)** is used to generate the vector $b(\alpha)=\{0\}\cup \{  -|t_j|^{\alpha}\ln(|t_j|)\exp(-|t_j|^{\alpha}),\quad j=1,2,..,k \}$. The usage of these functions enables us to formulate the following equation 
$$\sum_{v=1}^k \color{red}{a_v} \color{black}\cdot (\exp\left\{ -|t_v-t_j|^a \right\}+\exp\left\{ -|t_v+t_j|^a \right\})=-|t_j|^a \ln |t_j|\exp\left( -|t_j|^a\right)$$
for $j=1,2,...,k$, which is solved in terms of $a_v$, i.e. if $a(\alpha)=\{a_v,\quad v=1,2,...,k \}$, then $$a(\alpha)={\left( A(\alpha)\right)}^{-1} b(\alpha)$$
-  **F(X, T)** is used to calculate $$\sum_{j=1}^{n}F(\mathbf{T}, X_j)=\left\{2, 2\sum_{j=1}^N\cos(t_1 X_j), 2\sum_{j=1}^N\cos(t_2 X_j), ..., 2\sum_{j=1}^N\cos(t_k X_j) \right\}=2\psi_{\mathbf{X}}(\mathbf{T}).$$
The inputs of this function are as follows: $X$ is the vector of observations and $\mathbf{T}$ is the vector of the $t_j,\quad j=1,2,...,\mathrm{len}(\mathbf{T})$.
-  **projection(FX, T, a)** is used to calculate 
$$\left( (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top} \frac{1}{n}\sum_{j=1}^{n}F(\mathbf{T}, X_j)$$
-  **fisher_info(T, a)** is used to calculate the approximation of the Fisher information of the parameter $\alpha$: if $k$ is the length of $\mathbf{T}$, then $$I_k(\alpha) = \left( (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top} \cdot b(\alpha),$$ and $$\lim_{k \longrightarrow \infty} I_k(\alpha) = I(\alpha).$$

```python
def A(a, T):
    A, B = np.meshgrid(T, T)
    matrix = np.exp(-abs(A + B)**a) + np.exp(-abs(A - B)**a)
    return matrix

def B(a, T):  
    T = np.array(T, dtype=np.float64)
    T[T == 0] = np.finfo(np.float64).eps
    y = -abs(T)**a * np.log(abs(T)) * np.exp(-abs(T)**a)
    return y

def F(X, T):
    X1 = np.outer(X, T)
    Y1 = np.cos(X1)
    FX = np.mean(Y1, axis=0)
    return 2*FX

def projection(FX, T, a):
    return abs(np.dot(np.dot(np.linalg.inv(A(a, T)), B(a, T)), FX))
    

def fisher_info(T, a):
    return 2*abs(np.dot(np.dot(np.linalg.inv(A(a, T)), B(a, T)), B(a, T)))    
```

# **Generating samples from Stable distributions**

A sample from $S_{\alpha}(1,\beta,0)$ can be generated using the following formula

$$X = S_{\alpha, \beta}\cdot \frac{\sin(\alpha(V+B_{\alpha, \beta}))}{\cos(V)^{\alpha}} {\left(  \frac{ \cos(V-\alpha (V+B_{\alpha, \beta})) }{W} \right)}^{ \frac{1-\alpha}{\alpha} },$$


where $W\sim \exp(1)$, $V\sim U\left(-\frac{\pi}{2}, \frac{\pi}{2} \right)$, $V$ and $W$ are independent, and
\begin{align*}
B_{\alpha, \beta}&=\frac{\mathrm{arctan} \left( \beta \tan \frac{\pi \alpha}{2}\right) }{\alpha} \\
S_{\alpha, \beta}&={\left( 1 + {\left( \beta \tan\left( \frac{\pi \alpha}{2} \right) \right)}^2 \right)}^{\frac{1}{2\alpha}}
\end{align*} 

```python
#Generation of samples from stable distribution

def beta(a, b):
    return np.arctan(b*np.tan(np.pi*a/2))/a

def svalue(alpha, bett):
    val_beta = bett*bett
    val_tan  = np.tan(np.pi*alpha/2)*np.tan(np.pi*alpha/2)
    power_val = 1/(2*alpha)
    return (1+val_beta*val_tan)**power_val

def stable_sample(alpha, N):
    bt = 0
    sample_exp = np.random.exponential(scale=1, size=N)
    sample_uni = np.random.uniform(low=-np.pi/2, high=np.pi/2, size=N)
    Y1 = svalue(alpha, bt)*np.sin(alpha*(sample_uni + beta(alpha, bt)))
    Y2 = np.cos(sample_uni)
    Y2 = Y2**(-1/alpha)
    Y3 = np.cos(sample_uni - alpha*(sample_uni + beta(alpha, bt)))
    Y3 = Y3**((1-alpha)/alpha)
    Y4 = sample_exp**((alpha-1)/alpha)
    Y = Y1*Y2
    Y *= Y3
    Y *= Y4
    return Y
```

The following function demonstrates the estimation of $\alpha$ with the methodology of projections. A user defines a list of the value of $\alpha$ and this function provides estimates of those values of the tail index.


The number of those estimates is determined by the ***input N_samples***. These estimates are aggregated, and the average $\mu$ of those estimates and the $\mu \pm 3\sigma$ intervals are visualized. 

If all the estimates of $\alpha$ from that picture lie on the line $y(x) = x$, the estimate is considered perfect. The deviations from this line measure the error of the estimates. In the output plot, the true values of $\alpha$ are represented on the line $y(x)=x$.


```python
def calculate_estimates_and_visualize(N, gran, N_samples, start_alpha, end_alpha, portfolio_of_functions, N_f, verbose):
    # Description of the input variables
    # N                       ... sample size
    # gran                    ... the number of alpha values that are estimated and tested
    # N_samples               ... the number of samples for each value of alpha
    # start_alpha             ... the first (smallest) alpha (but bigger than 0.1)
    # end_alpha               ... the last  (biggest)  alpha (but not larger than 2)
    # portfolio_of_functions  ... the set functions each of which can be used to build T
    # N_f                     ... the index of the function in portfolio_of_function that will be applied
    # verbose                 ... if it is one, it will print intermediary results and visualize the final output

    KEY = list(portfolio_of_functions.keys())
    Means = []
    STDs = []
    values_of_alpha = np.linspace(start_alpha, end_alpha, gran)
    Fishr_invers = []
    for xa in values_of_alpha:
        # Forming the array of the estimates of xa
        estimates = []
        for i in range(N_samples):
            fun = KEY[N_f]
            # Generation of a sample from S_{\alpha}(1, 0, 0) with sample size N
            X = stable_sample(xa, N)
            # Production of the vector T for functions A(a, T), b(a, T), etc.
               # Production of the initial interval with equidistant points
            T = np.arange(0, 3, 0.05)
            # The application of the function of choice on T
            T = portfolio_of_functions[fun](T)
            # Making sure that at least one element is close to unit in array T
            if max(T) < 0.999:
               T[-1] = 0.999
            FX = F(X, T)
            fun0 = lambda x: projection(FX, T, x)
            alpha_est = extended_binary_search(fun0, start_alpha, end_alpha)
            # Additing the estimate to the array
            estimates.append(alpha_est)
        # if you prefer to see the intermediate results you can choose verbose to be equal to 1.    
        if verbose:
          print('alpha = '+truncate_string(xa)+', est = '+truncate_string(np.mean(estimates))+', std = '+truncate_string(np.std(estimates)))
        # The forming of the arrays of the means and STDs of the estimates
        Means.append(np.mean(estimates))
        STDs.append(np.std(estimates))
        Fishr_invers.append(1/np.sqrt(fisher_info(T, xa)))

    # The calculation of the average and maximum bias of the estimates
    Bias_avg = np.mean((np.array(Means) - np.array(values_of_alpha))**2)
    Bias_max = max((np.array(Means) - np.array(values_of_alpha))**2)
    # The calculation of the average and maximum std of the estimates
    Std_avg = np.mean(STDs)
    Std_max = max(STDs)
    if N < 200:
      print('---------------------------------------------------------------')
      print('N = '+str(N)+'.   For fun(x) = '+fun+' avg_max_bias = '+truncate_string(Bias_avg)+    ', min_max_bias = '+truncate_string(Bias_max))
      print('N = '+str(N)+'.   For fun(x) = '+fun+' avg_max_std  = '+truncate_string(Std_avg)+', min_max_std  = '+truncate_string(Std_max))
    else:
      print('---------------------------------------------------------------')
      print('N = '+str(N)+'. For fun(x) = '+fun+' avg_max_bias = '+truncate_string(Bias_avg)+    ', min_max_bias = '+truncate_string(Bias_max))
      print('N = '+str(N)+'. For fun(x) = '+fun+' avg_max_std  = '+truncate_string(Std_avg)+', min_max_std  = '+truncate_string(Std_max))
    # The calculation of mu plus/minus 3sigma
    upper_line = np.array(Means) + 3*np.array(STDs)
    lower_line = np.array(Means) - 3*np.array(STDs)
    # Visualization
    if verbose:
        fig, ax = plt.subplots(figsize=(14, 8), ncols=2)

        ax[0].plot(values_of_alpha, Means, 'g-o')
        ax[0].plot(values_of_alpha, upper_line, 'b')
        ax[0].plot(values_of_alpha, lower_line, 'r')
        ax[0].plot(values_of_alpha, values_of_alpha, 'k')
        ax[0].fill_between(values_of_alpha, upper_line, lower_line, color='gray', alpha=0.5)
        ax[0].grid()
        ax[0].set_xlabel(r'$\alpha$')
        ax[0].set_ylabel(r'estimate $\alpha$')
        ax[0].set_title(r'Average estimates and their $\mu \pm 3 \sigma$ intervals. N = ' + str(N))
        ax[0].legend(['$\mu$,  $\mu = avg(alpha\_sample)$','$ \mu + 3*\sigma$,  $\sigma = std(alpha\_sample)$', '$\mu - 3*\sigma$',  'real_alpha'])

        # Your original code for the right subplot
        ax[1].plot(values_of_alpha, np.array(STDs)*np.sqrt(N), 'g-o')
        ax[1].plot(values_of_alpha, Fishr_invers)
        ax[1].set_xlabel(r'$\alpha$')
        ax[1].set_ylabel(r'$1/\sqrt{FisherInfo(\alpha)}$')
        ax[1].grid()
        ax[1].legend([r'$\sqrt{N}$*STD','Inverse_fisher_info'])
        ax[1].set_title(r'STD*$\sqrt{N}$, N = '+str(N))

        # Adjust the spacing between the subplots
        plt.subplots_adjust(wspace=0.3)
    return Means, STDs, Bias_max, Std_max
```

Let us recall that
$$J_k(\mathbf{X}, \mathbf{T}, \alpha) =  \left( (A(\alpha))^{-1}\cdot b(\alpha) \right)^{\top} \frac{1}{n}\sum_{m=1}^{n}F(\mathbf{T}, X_m),$$

and let us fix the vector of observations $\mathbf{X}$ and the vector $\mathbf{T}$ and define 
$$\mathrm{Error}(\alpha) = \left|J_k(\mathbf{X}, \mathbf{T}, \alpha)\right|.$$

Hence, $\hat{\alpha}_{_{MLP}}=\mathrm{argmin}_{a} \mathrm{Error}(a)$ is the MLP estimate of $\alpha$.

In the following cells, the function $\mathrm{Error}(a)$ will be visualized for various $X$ and various true values of $\alpha$. 

For some inputs, $\mathrm{Error}(a)$ is unimodal in $a$ and binary search can be applied. For other choices, the function is not unimodal and binary search can result in incorrect estimates. This issue is addressed by the extended binary search.

```python
# The following function looks unimodal and binary search can be applied
test_alpha = 1.4
N = 8000
start_alpha = 0.5 
end_alpha = 1.9
gran_of_choice = 0.01

# Simulation of a sample
X = stable_sample(test_alpha, N)
estimates = []
T = np.arange(0, 0.5, 0.07)
T = portfolio_of_functions[fun](T)
if max(T)<0.999:
  T[-1] = 0.999
Alpha = list(np.arange(start_alpha, end_alpha, gran_of_choice))+[]
FX = F(X, T)
crit = []
for al in Alpha:
    y = projection(FX, T, al)
    crit.append(y)

fun0 = lambda x: projection(FX, T, x)

argmin = np.argmin(crit) 
alpha_bin_search = extended_binary_search(fun0, start_alpha, end_alpha)
alpha_straighforward = Alpha[argmin]
print('alpha = '+str(round(test_alpha, 4))+', alpha_straightforward_search = '+str(round(alpha_straighforward, 4))+', alpha_extended_bin_search = '+str(round(alpha_bin_search, 4)))
print('-------------------------------------------------------------------')

fig = plt.figure(figsize=(8, 6))
plt.plot(Alpha, crit)
plt.grid()
plt.xlabel('alpha: Error(alpha) is a uni-modal function')
plt.ylabel('Error(alpha)')
plt.title('Binary search can be applied: alpha = '+str(test_alpha)+', alpha_bin_search = '+str(round(alpha_bin_search,4)))
plt.axvline(x=test_alpha, color='red', linestyle='--')
plt.axvline(x=alpha_bin_search, color='green', linestyle='--')
plt.show()
```

Let us recall that the a by-product of this methodology is an algorithm of a numerical calculation of the Fisher information: $$I_k(\alpha, \mathbf{T}) = \left( (A(\alpha, \mathbf{T}))^{-1}\cdot b(\alpha, \mathbf{T}) \right)^{\top} \cdot b(\alpha, \mathbf{T}),$$ and $I(\alpha, \mathbf{T})$ converges to $I(\alpha).$


Finding the optimal $\mathbf{T}$ can be achieved by methods based on hyper-parameter selection, e.g. grid search, random search, etc. Since the matrix $A(\alpha, \mathbf{T})$ must be inverted and its dimension is $\mathrm{length}(\mathbf{T})\times \mathrm{length}(\mathbf{T})$, larger dimension of $\mathbf{T}$ does not necessarily imply that the considered estimate will be better: the larger the matrix, the more likely are rounding errors.

In this report, the vector $\mathbf{T}$ is determined as follows: 



1.   Take a set of equidistant points: 
$\mathbf{T}_0 = \{0, \delta, 2\cdot \delta, \dots ,n\cdot \delta\}$ where $n$ is the maximum natural number for which $n\cdot \delta \leq m$
2.   Apply $\mathbf{f}$ on the vector $\mathbf{T}_0$, i.e. $\mathbf{T}=\mathbf{f}(\mathbf{T}_0)$.

In other words, $\mathbf{T}$ is a function of $\delta$, $m$, and $\mathbf{f}$. The grid search can be conducted over these arguments. The set of candidate functions in this report is in the dictionary variable **portfolio_of_functions**.

In this report a grid search was conducted with manual editings, and the following parameters of $\mathbf{T}$ were used: $\delta = 0.05$, $m = 3$, and $\mathbf{f}(x) = \min(x^3, x^5)$.

Anyway, there is enoug room to improve the demonstrated results. 


Note that $S_{\alpha}(1,0,0)$ is a symmetric distribution, therefore the vector $\mathbf{T}$ should contain only nonnegative elements. Note that if $\mathbf{T}$ contains $t_j > 0$, and $-t_j$, then the matrix $A(\alpha, \mathbf{T})$ can become singular. This is not the case for a general stable distribution.

The following function is the criterion for the grid search. For two vectors of the same dimension, it ascertains if $X_1 > X_2$ which means that $X_1(i) \geq X_2(i)$ for all $i\in \Xi = \{1,2,..., \mathrm{len}(X1)\}$, and there exists at least one $j \in \Xi$ such that $X_1(j) > X_2(j)$.

```python
# Check if X1 >= X2 element-wise
def comparison_of_vectors(X1, X2):
    if np.all(X1 >= X2):
        larger_or_equal = 1
    else:
        larger_or_equal = 0

    # Check if there exists at least one index j such that X1[j] > X2[j]
    if np.any(X1 > X2):
        strictly_larger = 1
    else:
        strictly_larger = 0
    return larger_or_equal*strictly_larger
```

# **Appendix**

**Proof about the formula of calculating the Fisher information**


In the ***maximum likelihood projections*** method (MLP), the function $J(x,\alpha)$ is projected to the space $\{1, \exp(i\cdot t_1 x), \exp(i\cdot t_2 x), ..., \exp(i \cdot t_k x) \}$, i.e.
$$J_k(x,\alpha) = \sum_{j=0}^{k}a_j \exp(i\cdot t_jx),$$

Let us consider $J_k^2(x,\alpha)$ because $I(\alpha)=\mathbb{E}J^2(X,\alpha)=\lim_{k\rightarrow \infty}\mathbb{E}J_k^2(X,\alpha)$

Thus,
$$J^2_k(x,\alpha) = {\left(\sum_{j=0}^{k}a_j \exp(i\cdot t_jx)\right)}^2=\sum_{j=0}^{k}a_j^2 \exp(i\cdot t_j x)^2+\sum_{i\neq j}a_i a_j \exp(i \cdot t_i)\exp(i \cdot t_j)=$$
$$=\sum_{j=0}^k a_j^2 \exp(i\cdot 2\cdot t_j \cdot x)+\sum_{i\neq j}a_i a_j \exp(i\cdot (t_i+t_j)\cdot x).$$

Hence, 

\begin{align}
\mathbf{(1)}\quad \mathbb{E}J^2_k(X,\alpha)&=\sum_{j=0}^k a_j^2 \mathbb{E}\exp(i\cdot 2\cdot t_j \cdot X)+\sum_{i\neq j}a_i a_j \mathbb{E}\exp(i\cdot (t_i+t_j)\cdot X)=\\
&=\sum_{j=0}^k a_j^2 \psi(2t_j)+\sum_{i\neq j}a_i a_j \psi(t_i+t_j).
\end{align}


Let us remind that 

$$(a_0,a_1,...,a_k)=\left(\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top}.$$

Hence 

\begin{align*}
\mathrm{Fisher\_info}(\mathbf{T},\alpha) & = \sum_{j=0}^k a_j^2 \psi(2t_j)+\sum_{i\neq j}a_i a_j \psi(t_i+t_j)=\\
&= \left(\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top} \psi(\mathbf{T}{\oplus}\mathbf{T},\alpha) \psi(\mathbf{T}{\oplus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}=\\
&= \left(\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top}\frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha},
\end{align*}


which proves the statement.

# **Empirical Fisher Information**

In this report, a parallel notation is used and
$$a(\alpha)^\top =  (a_0,a_1,...,a_k)=\left(\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top}.$$

According to $\mathbf{(1)}$, $\mathbb{E}J^2_k(X,\alpha)=\sum_{j=0}^k a_j^2 \psi(2t_j)+\sum_{i\neq j}a_i a_j \psi(t_i+t_j).$

It can easily be shown that if $\mathbf{X}=(X_1,...,X_n)$ is a sample from $X\sim \psi(t)$, then
$$\frac{1}{n}\sum_{j=1}^N J_k^2(X_j,\alpha)=\sum_{j=0}^k a_j^2 \psi_{\mathbf{X}}(2t_j)+\sum_{i\neq j}a_i a_j \psi_{\mathbf{X}}(t_i+t_j).$$

Taking into account that
$$\mathrm{Fisher\_info}(\mathbf{T},\alpha)=\left(\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top} \color{red}\psi\color{red}(\mathbf{\color{red}T}{\color{red}\oplus}\mathbf{\color{red}T}\color{red},\color{red}\alpha\color{red}) \psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha},$$
and that the matrix highlighted in red will disappear due to its neighbor matrix on the right, we get the expression of $\mathrm{Fisher\_info}(\mathbf{T},\alpha)$. In the empirical case, however, the matrix highlighted in red is the application of the empirical characteristic function on matrix $\mathbf{T}\oplus \mathbf{T}$, i.e.:

$$\mathrm{Emp\_FI}_{\mathbf{X}}(\mathbf{T},\alpha)=\left(\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top} \color{red}\psi_{\mathbf{X}}\color{red}(\mathbf{\color{red}T}{\color{red}\oplus}\mathbf{\color{red}T}\color{red},\color{red}\alpha\color{red}) \psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}.$$

This can be proved in the following way:
$$\sum_{j=0}^k a_j^2 \psi_{\mathbf{X}}(2t_j)+\sum_{i\neq j}a_i a_j \psi_{\mathbf{X}}(t_i+t_j)=a(\alpha)^\top \psi_{\mathbf{X}}(\mathbf{T}\oplus \mathbf{T})a(\alpha).$$
Hence, by taking into account the fact that $a(\alpha)^\top=\left(\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top}$, we finalize the proof.

# **Optimization problem**

The aforementioned values $\hat{\alpha}_{_{MLP}}(\mathbf{T})$ and $\mathrm{Emp\_FI}_{\mathbf{X}}(\mathbf{T},\hat{\alpha}_{_{MLP}}(\mathbf{T})) $ are a numerical solution of the following optimization problem

\begin{align}
\mathbf{(P)}\quad \max_{\alpha} &\quad \frac{1}{n}\sum_{j=1}^{n} J^2(X_j,\alpha)\\
\mathrm{s.t.} &\quad J(x,\alpha)=\frac{  \left( \frac{\partial p(x,\alpha)}{\partial \alpha} \right) }{p(x,\alpha)} \\
&\quad p(x,\alpha) = \frac{1}{2\pi}\int_{\mathbb{R}}\exp(-i t x)\psi(t,\alpha)dt,
\end{align}
given a sample $\mathbf{X}=(X_1,X_2,\dots,X_n)$.


Using the insights from this report, it can easily be shown that $$\lim_{\mathbf{T}\rightarrow\mathbf{T}_\psi}\hat{\alpha}_{_{MLP}}(\mathbf{T})=\mathrm{argmax}\mathbf{(P)}\quad \text{and}\quad \lim_{\mathbf{T}\rightarrow\mathbf{T}_\psi}\mathrm{Emp\_FI}_{\mathbf{X}}(\mathbf{T},\hat{\alpha}_{_{MLP}}(\mathbf{T}))=\mathrm{max}\mathbf{(P)}.$$

# **The Fisher information for the symmetric case**

Let us recall that for the symmetric case, the function $J_k$, the MLP estimator and the vector of projections are of the following form:

\begin{align*}
J^*_k(x,\alpha) &= 2a^*_0 + 2\sum_{j=1}^{k}a_j\cos(t_j x) \\
\hat{\alpha}_{_{MLP}}& =\operatorname{agrmin}_{\alpha}\left| \left(  (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)+\psi(\mathbf{T}{\ominus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} \right)^{\top} \cdot \psi_{\mathbf{X}}(\mathbf{T}) \right| \\
a(\alpha,\mathbf{T}) &=(a_0,a_1,a_2,...,a_k)^\top = (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)+\psi(\mathbf{T}{\ominus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha} 
\end{align*}

Taking these facts into account, let us calculate $J_k^2(x,\alpha)$:
\begin{align}
J^2_k(x,\alpha)&={\left(\sum_{j=0}^{k}2 a_j \cos(i\cdot t_jx)\right)}^2 =\\
& = {\left(\sum_{j=0}^{k}a_j \left( \exp(i\cdot t_jx) + \exp(-i\cdot t_jx) \right) \right)}^2 = \\
& = {\left(   \sum_{j=0}^{k}a_j \exp(i\cdot t_jx) + \sum_{j=0}^{k}a_j \exp(-i\cdot t_jx)   \right)}^2 = \\
&=  {\left(   \sum_{j=0}^{k}a_j \exp(i\cdot t_jx)  \right)}^2 + {\left(   \sum_{j=0}^{k}a_j \exp(-i\cdot t_jx)  \right)}^2 + \\
&+2\cdot \sum_{j=0}^{k}a_j \exp(i\cdot t_jx) \sum_{j=0}^{k}a_j \exp(-i\cdot t_jx) 
\end{align}

Our goal is to calculate $\mathbb{E}J^*_k(X,\alpha)$ which is the MLP approximation of the Fisher information. Hence, it follows:

\begin{align}
\mathbb{E}J^2_k(X,\alpha)&=  \mathbb{E}{\left(   \sum_{j=0}^{k}a_j \exp(i\cdot t_j X)  \right)}^2 + \mathbb{E}{\left(   \sum_{j=0}^{k}a_j \exp(-i\cdot t_j X)  \right)}^2 + \\
&+2\mathbb{E}\cdot \sum_{j=0}^{k}a_j \exp(i\cdot t_j X) \sum_{j=0}^{k}a_j \exp(-i\cdot t_j X) 
\end{align}
Since, $X$ is symmetric, $\mathbb{E}\exp(i\cdot t \cdot X) = \mathbb{E}\exp(-i\cdot t \cdot X)$, and it can easily be shown that 
$$\mathbb{E}{\left(   \sum_{j=0}^{k}a_j \exp(i\cdot t_j X)  \right)}^2 = \mathbb{E}{\left(   \sum_{j=0}^{k}a_j \exp(-i\cdot t_j X)  \right)}^2.$$
Hence we get:
\begin{align}
\mathbb{E}J^2_k(X,\alpha)&= 2 \mathbb{E}{\left(   \sum_{j=0}^{k}a_j \exp(i\cdot t_j X)  \right)}^2  + \\
&+2\mathbb{E}\cdot \sum_{j=0}^{k}a_j \exp(i\cdot t_j X) \sum_{j=0}^{k}a_j \exp(-i\cdot t_j X) 
\end{align}

From the proof for the general case, we know that
$$\mathbb{E}{\left(\sum_{j=0}^{k}a_j \exp(i\cdot t_j X)\right)}^2=\sum_{j=0}^k a_j^2 \psi(2\cdot t_j )+\sum_{i\neq j}a_i a_j \psi(t_i+t_j).$$

The second summand equals

$$\mathbb{E}\cdot \sum_{j=0}^{k}a_j \exp(i\cdot t_j X) \sum_{j=0}^{k}a_j \exp(-i\cdot t_j X)=\sum_{j=0}^k a_j^2 \psi(0)+\sum_{i\neq j}a_i a_j \psi(t_i-t_j)$$

Hence:

\begin{align}
\mathbb{E}J^2_k(X,\alpha)&= \mathbf{2}\sum_{i=0}^k \sum_{j=0}^k a_i a_j \left(  \psi(t_i+t_j) + \psi(t_i-t_j) \right) \\
&= \mathbf{2}\cdot a(\alpha,\mathbf{T})^\top \left( \psi(\mathbf{T} \oplus \mathbf{T}) + \psi(\mathbf{T}\ominus \mathbf{T}) \right) a(\alpha,\mathbf{T})\\
&= \mathbf{2} {\left( (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)+\psi(\mathbf{T}{\ominus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}\right)}^\top\times \\
& \times \left( \psi(\mathbf{T} \oplus \mathbf{T}) + \psi(\mathbf{T}\ominus \mathbf{T}) \right)\cdot   (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)+\psi(\mathbf{T}{\ominus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}=\\
&=\mathbf{2} {\left( (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)+\psi(\mathbf{T}{\ominus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}\right)}^\top \cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}.
\end{align}

Hence, the following theorem is proven:

**Theorem.** 
If $X\sim \psi(t,\alpha)$ is a symmetric centered random variable, i.e. symmetric with respect to zero, with parameter $\alpha$, and the following conditions are satisfied:


*   Its $J$ function is approximated as follows: $J^*_k(x,\alpha) = 2a^*_0 + 2\sum_{j=1}^{k}a_j\cos(t_j x)$
*   $\mathbf{T}$ is a discrete finite subset of $\mathbf{T}_\psi$.

Then
$$\mathrm{Fisher\_info}(\mathbf{T},\alpha)=\mathbf{\color{red} 2} {\left( (\psi(\mathbf{T}{\oplus} \mathbf{T},\alpha)+\psi(\mathbf{T}{\ominus} \mathbf{T},\alpha))^{-1}\cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}\right)}^\top \cdot \frac{\partial \psi(\mathbf{T}, \alpha)}{\partial \alpha}.$$



We emphasize that the 2 cannot be forgotten, therefore it is in bold font.
